# Elastoplasticity: Isotropic and kinematic hardening

### Generalized Rheological Model  

In this tutorial, we use a **generalized rheological model** capable of simultaneously capturing both **linear isotropic hardening** and **linear kinematic hardening**. This generalized model makes things easier as it includes the cases of only isotropic hardening and only kinematic hardening as well as ideal elastoplasticity as special cases. The analysis is conducted within the **geometrically linear regime**, i.e. **small strains**.

#### **Rheological Model Description**  

The model, depicted below, consists of **three basic rheological elements** and involves **four material parameters**:  

![Rheological Model](rheological-model.jpeg)


1. **Spring element with elasticity modulus $E$**  

   - Represents the initial purely **elastic response** of the material.  



2. **Friction element**  

   - Characterizes **isotropic hardening**, where the yield stress $\sigma_y$ evolves according to:  

     $$\sigma_y = \sigma_{y0} + H \kappa$$  

     where:  

     - $\sigma_{y0}$ is the **initial yield stress**,  
     - $H$ is the **isotropic hardening modulus**,  
     - $\kappa$ is the **isotropic hardening internal variable**, also known as the accumulated plastic strain, given by:  

       $$\kappa = \sum_i |(\epsilon_{p})_i|$$  

     - As $\sigma_y$ is given by a linear function, the isotropic hardening is said to be linear.  
     - If $H = 0$, **no isotropic hardening** occurs.  


3. **Spring Element with elasticity modulus $E_2$**  

   - Responsible for **(linear) kinematic hardening**.  
   - If $E_2 = 0$, there will be **no kinematic hardening**.  


#### **Loading Conditions**  
The material is subjected to **strain-controlled** loading, which means that we perscribe how the strain is applied, and accordingly calculate the corresponding stress based on the material law. The loading is assumed to be **quasi-static**, that is, it is applied sufficiently slowly that dynamic effects could be omitted.


# Example 1: Ideal elastoplasticity

In the first example, we set both $H$ and $E_2$ to zero. In result, the material yields with no hardening (i.e. ideal elastoplasticity). We set $E_1 = 200$ GPa and $\sigma_{y0} = 250$ MPa. In this case, we observe no change in the yield surface in terms of its size and position.

```python
# Import the solver from /src directiory (Don't touch!)
import sys
import os
project_path = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(os.path.join(project_path, "src"))
from elastoplasticity.elastoplasticity import ElastoplasticSolver, ElastoplasticMaterial, LinearInterpolator

# Import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Perscribe applied strin over time
t_data = np.array([0, 5, 10, 15]) # Discrete points in time of the strains to be interpolated
eps_data = np.array([0, 0.007, 0, 0.007]) # Strains to be interpolated
N = 10000 # To how many steps each interval is divided?

# Define material parameters in MPa
E_1 = 200000
E_2 = 0 # No kinematic hardening!
sigma_y0 = 250
H = 0 # No isotropic hardening!

# Create objects
interpolator = LinearInterpolator(t_data, eps_data, N)
material = ElastoplasticMaterial(E_1, E_2, sigma_y0, H)
solver = ElastoplasticSolver(material, interpolator)

# Compute results
t, eps, sigma = solver.solve()

# Plot results
plt.figure(figsize=(12, 5))

# Applied strain graph
plt.subplot(1, 2, 1)
plt.plot(t, eps, 'k-', linewidth=2)
plt.xlabel('Time [s]')
plt.ylabel('Strain [-]')
plt.title('Applied strain')
plt.grid()

# Stress-Strain diagram
plt.subplot(1, 2, 2)
plt.plot(eps, sigma, 'k-', linewidth=2)
plt.xlabel('Strain [-]')
plt.ylabel('Stress [MPa]')
plt.title('Stress-Strain diagram')
plt.grid()

plt.show()
```

# Example 4: Comparison between isotropic and kinematic hardening
In this example, we compare the two limiting cases of isotropic and kinematic hardening. The difference between them is clear; isotropic hardening (in blue) involves an expansion in the yield surface without shifting it, while kinematic hardening (in red) involves shifting the yield surface without expanding it.

```python
# Import the solver from /src directiory (Don't touch!)
import sys
import os
project_path = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(os.path.join(project_path, "src"))
from elastoplasticity.elastoplasticity import ElastoplasticSolver, ElastoplasticMaterial, LinearInterpolator

# Import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Given time and strain data
t_data = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45]) # Discrete points in time of the strains to be interpolated
eps_data = np.array([0, 0.012, -0.008, 0.018, -0.012, 0.025, -0.015, 0.032, -0.018, 0.005]) # Strains to be interpolated
N = 10000 # To how many steps each interval is divided?

# Define material parameters for purely isotropic hardening
E_1 = 200000  # MPa
E_2_iso = 0   # MPa
sigma_y0 = 250  # MPa
H_iso = 5000  # MPa

# Define material parameters for purely kinematic hardening
E_2_kin = 10000  # MPa
H_kin = 0  # MPa

# Create objects for isotropic hardening
interpolator_iso = LinearInterpolator(t_data, eps_data, N)
material_iso = ElastoplasticMaterial(E_1, E_2_iso, sigma_y0, H_iso)
solver_iso = ElastoplasticSolver(material_iso, interpolator_iso)

# Create objects for kinematic hardening
interpolator_kin = LinearInterpolator(t_data, eps_data, N)
material_kin = ElastoplasticMaterial(E_1, E_2_kin, sigma_y0, H_kin)
solver_kin = ElastoplasticSolver(material_kin, interpolator_kin)

# Compute results
t_iso, eps_iso, sigma_iso = solver_iso.solve()
t_kin, eps_kin, sigma_kin = solver_kin.solve()

# Plot results
plt.figure(figsize=(12, 5))

# Strain-Time graph
plt.subplot(1, 2, 1)
plt.plot(t_iso, eps_iso, 'k-', linewidth=2)
plt.xlabel('Time [s]')
plt.ylabel('Strain [-]')
plt.title('Applied strain')
plt.grid()


# Stress-Strain graph
plt.subplot(1, 2, 2)
plt.plot(eps_iso, sigma_iso, 'b-', linewidth=2, label='Only isotropic Hardening')
plt.plot(eps_kin, sigma_kin, 'r-', linewidth=2, label='Only kinematic Hardening')
plt.xlabel('Strain [-]')
plt.ylabel('Stress [MPa]')
plt.title('Stress-Strain diagram')
plt.legend(loc="lower right")
plt.grid()

plt.show()
```
